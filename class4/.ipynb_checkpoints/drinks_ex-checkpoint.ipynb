```python
# renaming one or more columns
drinks.rename(columns={'total_litres_of_pure_alcohol': 'pure_alcohol'})
```

```python
# replace all column names
drinks.rename(columns={'beer_servings':'beer','spirit_servings': 'spirit','wine_servings':'wine','total_litres_of_pure_alcohol':'pure_alcohol'})
```

```python
# add a new column as a function of existing columns
drinks['drink_more_beer'] = (drinks.beer_servings +20)
drinks['total_servings'] = (drinks.beer_servings + drinks.spirit_servings + drinks.wine_servings)
drinks
```

```python
#Exercise 3
# for each continent, calculate the mean beer servings
bycontinent = drinks.groupby('continent')
bycontinent.mean()
```

```python
# for each continent, calculate the mean of all numeric columns
drinks.groupby('continent').mean()
```

```python
# for each continent, describe beer servings
drinks.groupby('continent').describe()
```

```python
# similar, but outputs a DataFrame and can be customized
drinks.groupby('continent').agg(['count','mean','std'])
```

```python
drinks.groupby('continent').describe()
```

```python
# for each continent, count the number of occurrences
drinks.groupby('continent').count()
```
